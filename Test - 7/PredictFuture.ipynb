```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv(r'^NSEI\1m till 2022-02-18.csv')
dataset
```

```python
dataset.drop(['Datetime', 'Adj Close', 'Volume'], axis=1, inplace=True)
dataset
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
dataset_scaled = sc.fit_transform(dataset)
dataset_scaled
```

```python
# Creating a data structure with 60 timesteps and 1 output
x_train = []
x_train2 = []
y_train = []
for i in range(30, int(len(dataset_scaled)*0.8)):
    x_train.append(dataset_scaled[i-30:i])
    y_train.append(dataset_scaled[i, 3])
    temp = []
    for j in range(i-30,i):
        temp.extend(dataset_scaled[j])
    x_train2.append(temp)
x_train, y_train, x_train2 = np.array(x_train), np.array(y_train), np.array(x_train2)
x_train.shape, y_train.shape, x_train2.shape
```

```python
# Creating a data structure with 60 timesteps and 1 output
x_test = []
x_test2 = []
y_test = []
for i in range(int(len(dataset_scaled)*0.8), len(dataset_scaled)):
    x_test.append(dataset_scaled[i-30:i])
    y_test.append(dataset_scaled[i, 3])
    temp = []
    for j in range(i-30,i):
        temp.extend(dataset_scaled[j])
    x_test2.append(temp)
x_test, y_test, x_test2 = np.array(x_test), np.array(y_test), np.array(x_test2)
x_test.shape, y_test.shape, x_test2.shape
```

```python
train_x = list()
train_y = list()
n_size = 5
for i in range((x_train2.shape[0]//n_size)-1):
    train_x.append(x_train2[i*n_size:i*n_size+n_size])
    train_y.append(y_train[i*n_size:i*n_size+n_size])
train_x, train_y = np.array(train_x), np.array(train_y)
train_x.shape, train_y.shape
```

```python
test_x = list()
test_y = list()
n_size = 5
for i in range((x_test2.shape[0]//n_size)-1):
    test_x.append(x_test2[i*n_size:i*n_size+n_size])
    test_y.append(y_test[i*n_size:i*n_size+n_size])
test_x, test_y = np.array(test_x), np.array(test_y)
test_x.shape, test_y.shape
```

```python
#train the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
#ascending order
for i in range(len(train_x)):
#descending order
#for i in range(len(train_x)-1,-1,-1):
    print("Training model for fold", i+1)
    model.fit(train_x[i], train_y[i])
#model.fit(train_x, train_y)
```

```python
#Saving the model
import pickle
classifier = open('classifier.model', 'wb')
pickle.dump(model, classifier)
classifier.close()
```

```python
#Loading saved model
classifier = open('classifier.model', 'rb')
model2 = pickle.load(classifier)
classifier.close()
```

```python
#Testing the model
sos = 0
for i in range(len(test_x)):
    score = model2.score(test_x[i], test_y[i])
    sos = sos + score**2
rms = (sos / len(test_x)) ** 0.5
print("Accuracy RMS:", rms)
```
